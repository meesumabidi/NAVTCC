```python
import pandas as pd
import numpy as np
from sklearn import linear_model
```

```python
data = pd.read_excel('/content/House price data .xlsx')
data
```

```python
num_of_bedrooms = data['number of bedrooms']
area = data['Area of the house(excluding basement)']
condition= data['condition of the house']
data_set = pd.DataFrame([num_of_bedrooms,area,condition])
df=data_set.T
df
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['number of bedrooms','Area of the house(excluding basement)','condition of the house']],data.Price)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
reg.predict([[4,3310,838000]])
```
